```python
#Un código constructor
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as plt
from statistics import * #para min y max
```

```python
autos = pd.read_csv("autos2.csv") #defino la variable con el documento que quiero cargar
```

```python
# Assuming your price column is named 'price'
maximo= autos.price.max()
minimo=autos.price.min()
rango = maximo - minimo
media=autos.price.mean()
mediana=autos.price.median()
moda=autos.price.mode()


print ("Estadistica descriptiva de la Fabrica 2 \n", "\n Media: ", media,
       "\n Mediana: ", mediana, "\n Moda: ", moda, "\n Maximo: ", maximo,
       "\n Minimo: ", minimo, "\n Rango: ", rango)
```
