```python
#Un código constructor
import pandas as pd #para leer el documento
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
import matplotlib.pyplot as plt
import numpy as np
from statistics import * #para min y max, desviacion estandar
import math #para el numero de intervalos
```

**1)Cargar el Dataset de Autos y realizar las Estadísticas descriptivas básicas y medidas de tendencia central**

*Determinar el valor mínimo de precios*

```python
autos = pd.read_csv("autos2.csv" , encoding = "ISO-8859-1")#defino la variable con el documento que quiero cargar
```

```python
# Coloco despues del putno pa palabra price paraidentificar la columna con la
#que estoy trabajando

minimo=autos.price.min()
maximo= autos.price.max()

moda=mode(autos.price)
media= np.mean(autos.price)
mediana=median(autos.price)

desviacion=np.std(autos.price)


print ("Estadistica descriptiva de la Fabrica 2 \n", "\n Moda: ", moda,  "\n Media: ", media,
       "\n Mediana: ", mediana,
       "\n Maximo: ", maximo,
       "\n Minimo: ", minimo, "\n Desviación estándar: ", desviacion)
```

```python
autos.price.describe()
```

**2) Realizar un Histograma de precios de los automóviles**
Determinar mediante la regla de sturges la cantidad optima de bins o separadores

```python
 #para el histograma primero calcaular la cantidad de bins

n = len(autos.price)
k = int(1 + math.log2(n))
print("Valor de n: ", n, " \nNumero de intervalos: ", k)




#En base al paso anterior puedo decir que para bins necesito 8
plt.hist(autos.price,bins=8,alpha=0.2, color="b", edgecolor="black")


plt.axvline(media,color="b",linestyle="dashed",linewidth=2,label=f"Media: {media:.2f}")
plt.axvline(mediana,color="r",linestyle="dashed",linewidth=2,label=f"Mediana: {mediana:.2f}")
plt.axvline(moda,color="y",linestyle="dashed",linewidth=2,label=f"Moda: {moda:.2f}")
plt.axvline(desviacion,color="black",linestyle="dashed",linewidth=2,label=f"Desviacion: {desviacion:.2f}")

#Poner titulos y encabezados adecuados
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.title("Produccion diaria de la fabrica 1")
plt.legend()
plt.show()
```

**Interpretación de los datos y conclusiones**

Este histograma muestra la distribución de los precios de los autos y ya que no es simetrico, se nos indica que los precios no están distribuidos de manera uniforme sin grandes sesgos.
Puedo comentar que la media de los precios del archivo etiquetado como "Autos" es de 13,207.13. La mediana de 18,150.00 y la moda 16,500.00 y anterior a la media se encuentra la desviación media que es de 7,927.27 pesos respectivamente.
Y ya que la media es menor que la mediana y la moda, significa que la distribución está sesgada a la izquierda (sesgo negativo).

Puedo concluir que este histograma me sirvió para visualemtne identificar que tienen sesgo negativo y que la distribución no es uniforme.


**3) Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)**

```python
sns.boxplot(y = "body-style", x = "price", data = autos) #invertí los valaores
#de x y de y para poder lograr un display horizontal
```

**Interpretación de los datos y conclusiones:**

Compara los precios de los autos según su tipo de carrocería.
Una caja más grande significa mayor variabilidad de precios dentro de ese estilo.
La presencia de puntos fuera de los bigotes indica que hay autos con precios anómalos (outliers).

Puedo concluir que además de servirme para identificar visual y rapidamente el estilo con mayor variedad de precios, en general, los autos deportivos y sedanes de lujo tienden a mostrar precios más altos en comparación con hatchbacks o station wagons.

**4) Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo tracción ruedas (drive wheels)**

Si un tipo de tracción tiene una caja más pequeña, significa que sus precios son más uniformes dentro de esa categoría.
Puedo concluir visualmente que la tracción en las cuatro ruedas (4wd) suele asociarse con autos más caros, mientras que los de tracción delantera (fwd) suelen ser más económicos.
