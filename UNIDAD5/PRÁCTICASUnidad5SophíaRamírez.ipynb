```python
from sklearn.linear_model import LinearRegression # modelo de regresion lineal
import numpy as np
# datos de entrada
periodos = [2018, 2019, 2020, 2021, 2022,2023]
inversion = np.array( [2, 3, 5, 4, 11, 5] ).reshape(-1, 1)
ganancias = [20, 25, 34, 30, 40, 31]
# crear y entrenar el modelo de regresin lineal
modelo = LinearRegression()
modelo.fit(inversion,ganancias)

#Predecir la ganancia para el ao 2024
ganancia_2024 = modelo.predict([[8]])

print("La ganacia esperada para el ao 2024 es de: ", ganancia_2024[0])
# coeficientes de la regresion
print("Coeficiente de la pendiente (m): ", modelo.coef_[0])
print("Intercepcion de la recta (b):", modelo.intercept_)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Definir tamao
plt.figure(figsize=(6, 4))

#Crear el diagrama de dispersin con Lnea de regresin
sns.set(style="whitegrid")
sns.regplot(x=inversion, y=ganancias, scatter=True, color='blue')
plt.title('Diagrama de Dispersin con Lnea de Regresin')
plt.xlabel('Inversin (millones)')
plt.ylabel('Ganancias (millones)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
#Datos de entrada
publicidad = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1)
ganancias =np.array( [20, 25, 34, 30, 40, 31] )

#Crear y entrenar el modelo de regresion Lineal
modelo = LinearRegression()
modelo.fit(publicidad, ganancias)

#Predecir la ganancia para una inversin de 8 en la  publicidad
inversion_prediccion=   np.array([[8]])
ganancia_predicha= modelo.predict(inversion_prediccion)

print("La ganancia esperada para una inversin de 8 en publicidad es", ganancia_predicha[0])
print("Valor de la intercepcin", modelo.intercept_)
print("Valor de la pendiente", modelo.coef_)

#Crear et diagrama de dispersin con la Linea de regreston
plt.figure(figsize=(6, 5))
plt.scatter(publicidad, ganancias, color='blue', label='Datos')
plt.plot(publicidad, modelo.predict(publicidad), color='red', label='Linea de regresin')
plt.scatter(inversion_prediccion, ganancia_predicha, color= 'green', marker='x', s=100,label='prediccion (8, {0:.2f})'.format(ganancia_predicha[0]))
plt.title('Pronostico de ganancias a partir de la inversin en publicidad')
plt.xlabel('Inversin en Publicidad (millones)')
plt.ylabel('Ganancias Anuales (millones)')
plt.legend()
plt.grid(True)
plt.show()
```

##Grupo 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
grupo1 = pd.read_csv("grupo1.csv")
grupo1.head()
```

```python
# Histograma
plt.figure(figsize=(6,4))
sns.histplot(grupo1["Calificacion"], bins=10, kde=True, color='skyblue')
plt.title("Grupo 1 - Histograma de Calificaciones")
plt.xlabel("Calificacion")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

```python
# Regresin lineal
X1 = grupo1[["HorasEstudio"]].values
y1 = grupo1["Calificacion"].values
modelo1 = LinearRegression().fit(X1, y1)
pred1 = modelo1.predict(X1)

# Parmetros
pendiente1 = modelo1.coef_[0]
intercepto1 = modelo1.intercept_
r1, _ = pearsonr(grupo1["HorasEstudio"], grupo1["Calificacion"])
r2_1 = r2_score(y1, pred1)

print(f"Ecuacin: y = {pendiente1:.2f}x + {intercepto1:.2f}")
print(f"Coeficiente de correlacin de Pearson (r): {r1:.4f}")
print(f"Coeficiente de determinacin (R): {r2_1:.4f}")
```

```python
# Dispersin + recta
plt.figure(figsize=(6,4))
sns.scatterplot(x=grupo1["HorasEstudio"], y=grupo1["Calificacion"])
plt.plot(grupo1["HorasEstudio"], pred1, color='red', label='Lnea de regresin')
plt.title("Grupo 1 - Regresin Lineal")
plt.xlabel("HorasEstudio")
plt.ylabel("Calificacion")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Coeficiente de correlacin
print(f"Coeficiente de correlacin (Pearson) para Grupo 1: r = {r1:.4f}")
```

**Explicar cul es la interpretacin del coeficiente de regresin en trminos de la
relacin entre el tiempo de estudio y las calificaciones finales?**

En este caso, el coeficiente de regresin indica que por cada hora extra que un estudiante dedica al estudio, su calificacin final tiende a incrementarse en aproximadamente 2.47 puntos. Esto refleja una influencia significativa y positiva del tiempo de estudio sobre el rendimiento acadmico.

**Cmo se interpreta el coeficiente de correlacin en trminos de la fuerza y direccin de la relacin entre las variables?**

El coeficiente de correlacin obtenido es 0.95, lo que seala una asociacin positiva y muy fuerte entre las horas de estudio y las calificaciones. En trminos prcticos, esto implica que, generalmente, los estudiantes que estudiaron ms lograron mejores resultados acadmicos, y esta tendencia se mantiene de manera consistente.

**El modelo de regresin lineal es estadsticamente significativo?**

S, el modelo presenta una alta significancia estadstica. El valor de R es 0.91, lo que indica que el modelo puede explicar el 91% de la variacin en las calificaciones finales a partir del tiempo de estudio. Esto demuestra que existe una relacin slida y confiable entre ambas variables.

**Qu tan confiable es el modelo de regresin para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?**

El modelo es altamente confiable. Dado el elevado valor de R y la fuerte correlacin observada, se pueden realizar predicciones precisas y fundamentadas sobre el desempeo acadmico de los estudiantes en funcin del tiempo que dedican a estudiar.
